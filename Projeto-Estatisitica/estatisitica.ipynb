#### Ajustando gráfico 
 
 - destacar os picos de 1-2 anos e o de 7 anos 

```python
plt.figure(figsize=(10, 6))  # Aumentar o tamanho da figura

# Ajustar o parâmetro bw_adjust (bandwidth) para controlar a suavização da curva
sns.kdeplot(df[df['Demissão'] == 'Sim']['Anos desde a última promoção'], 
            label='Demitidos', 
            fill=True, 
            color='#ff6666',
            bw_adjust=0.8,  # Reduzir para detectar mais variações (menos suavização)
            alpha=0.5)      # Maior transparência para ver sobreposições

sns.kdeplot(df[df['Demissão'] == 'Não']['Anos desde a última promoção'], 
            label='Não Demitidos', 
            fill=True, 
            color='#66b3ff',
            bw_adjust=0.6,  
            alpha=0.5)

# Adicionar linhas verticais nos pontos de interesse
plt.axvline(x=1.5, color='gray', linestyle='--', alpha=0.7, label='Pico 1-2 anos')
plt.axvline(x=7, color='gray', linestyle='--', alpha=0.7, label='Pico ~7 anos')

# Ajustar a escala do eixo x para focar na região de interesse
plt.xlim(-1, 16)  

# Adicionar grade para facilitar a leitura
plt.grid(True, linestyle='--', alpha=0.3)

plt.title('Distribuição de Densidade: Anos desde a Última Promoção', fontsize=14)
plt.xlabel('Anos desde a Última Promoção', fontsize=12)
plt.ylabel('Densidade', fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df[df['Demissão'] == 'Sim']['Nível de envolvimento com o trabalho'], 
           label='Demitidos', fill=True, color='#ff6666', alpha=0.5)
sns.kdeplot(df[df['Demissão'] == 'Não']['Nível de envolvimento com o trabalho'], 
           label='Não Demitidos', fill=True, color='#66b3ff', alpha=0.5)
plt.title('Distribuição de Densidade: Nível de Envolvimento com o Trabalho')
plt.xlabel('Nível de Envolvimento')
plt.ylabel('Densidade')
# Adicionar anotação com o valor da correlação
plt.annotate('Correlação: -0.13', xy=(0.05, 0.95), xycoords='axes fraction', 
             bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))
plt.grid(True, linestyle='--', alpha=0.3)
plt.legend()
plt.show()
```

- nível 4 (alto envolvimento): Ligeiramente mais comum em não demitidos

mesmo com a correlação fraca o gráfico sugere que funcionarios com niveis muito baixos de envolvimento (1) tem probabilidade um pouco maior de saírem, enquanto aqueles com alto envolvimento (3-4) tendem a permanecer
